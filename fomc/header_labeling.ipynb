```python
import requests
import chardet
from bs4 import BeautifulSoup

def fetch_soup(url):
    res = requests.get(url)
    res.raise_for_status()

    enc = chardet.detect(res.content)["encoding"] or "utf-8"
    html = res.content.decode(enc, errors="replace")

    return BeautifulSoup(html, "html.parser")
```

```python
def extract_header_candidates(soup):
    header_tags = soup.find_all(["h1", "h2", "h3", "strong"])
    results = []

    for tag in header_tags:
        text = tag.get_text(" ", strip=True)
        text = " ".join(text.split())

        if not text:
            continue
        if len(text.split()) < 2:   # ë„ˆë¬´ ì§§ìœ¼ë©´ ì œì™¸
            continue
        if len(text.split()) > 20:  # ë„ˆë¬´ ê¸¸ë©´ ì œì™¸
            continue

        lower = text.lower()
        if any(skip in lower for skip in ["federal reserve", "contact", "accessibility", "return to top"]):
            continue

        results.append(text)

    # ì¤‘ë³µ ì œê±°
    seen = set()
    final = []
    for r in results:
        if r not in seen:
            seen.add(r)
            final.append(r)

    return final
```

```python
def header_to_section_label(header):
    h = header.lower()
    for label, keys in SECTION_LABELS.items():
        for k in keys:
            if k in h:
                return label
    return None
```

```python
def extract_section_blocks(soup):
    headers = extract_header_candidates(soup)
    paragraphs = soup.find_all("p")

    blocks = []
    current_label = None

    for p in paragraphs:
        text = p.get_text(" ", strip=True)
        if not text:
            continue

        # ë¬¸ë‹¨ì´ í—¤ë”ê°€ ë˜ë©´ ë¼ë²¨ ë°”ë€œ
        for h in headers:
            if text.startswith(h):
                current_label = header_to_section_label(h)
                break
        else:
            # í—¤ë”ê°€ ì•„ë‹ˆë©´ ë¬¸ë‹¨ì„ í•´ë‹¹ ì„¹ì…˜ì— ì¶”ê°€
            blocks.append({
                "label": current_label,
                "text": text
            })

    return blocks
```

```python
import nltk
nltk.download("punkt")

def block_to_sentences(blocks):
    results = []
    for b in blocks:
        label = b["label"]
        para = b["text"]

        for sent in nltk.sent_tokenize(para):
            sent = sent.strip()
            if not sent:
                continue
            results.append((sent, label))

    return results
```

```python
import difflib
import requests
from bs4 import BeautifulSoup
import chardet
import nltk

df = pd.read_csv("FOMC_corpus_sentence_level.csv")

def fetch_raw_sentences(url):
    res = requests.get(url)
    enc = chardet.detect(res.content)["encoding"] or "utf-8"
    soup = BeautifulSoup(res.content.decode(enc, errors="replace"), "html.parser")

    paras = [p.get_text(" ", strip=True) for p in soup.find_all("p")]
    sents = []
    for p in paras:
        for s in nltk.sent_tokenize(p):
            sents.append(s)
    return sents

# í…ŒìŠ¤íŠ¸ ë¬¸ì„œ í•˜ë‚˜ ê³¨ë¼ ê²€ì‚¬
test_doc = df["doc_id"].iloc[0]
date = test_doc.replace("FOMC_", "")
url = f"https://www.federalreserve.gov/monetarypolicy/fomcminutes{date}.htm"

html_sents = fetch_raw_sentences(url)
csv_sents = df[df["doc_id"] == test_doc]["text"].tolist()

for cs in csv_sents[:10]:
    match = max(html_sents, key=lambda x: difflib.SequenceMatcher(None, cs, x).ratio())
    ratio = difflib.SequenceMatcher(None, cs, match).ratio()
    print("\nCSV:", cs)
    print("Best HTML:", match)
    print("Similarity:", round(ratio, 3))
```
