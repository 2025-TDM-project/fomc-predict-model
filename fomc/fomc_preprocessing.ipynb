```python
import os
import re
import json
import chardet
import requests
import pandas as pd
from bs4 import BeautifulSoup

import nltk
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
import spacy

# ----------------------------------------------------
# ÌôòÍ≤Ω Ï§ÄÎπÑ
# ----------------------------------------------------
nltk.download("punkt")
nltk.download("stopwords")

nlp = spacy.load("en_core_web_sm")
stop_words = set(stopwords.words("english"))


# ----------------------------------------------------
# 1) HTML Î°úÎìú (ÏµúÏã†/Í≥ºÍ±∞ Minutes Î™®Îëê ÏßÄÏõê)
# ----------------------------------------------------
def fetch_minutes_text(url):
    print(f"\nüîµ [FETCH] {url}")

    res = requests.get(url)
    detected = chardet.detect(res.content)
    encoding = detected["encoding"] or "utf-8"

    soup = BeautifulSoup(res.content.decode(encoding, errors="replace"), "html.parser")

    # ÏµúÏã† Íµ¨Ï°∞
    main = soup.find("div", id="article")
    if main:
        paragraphs = main.find_all("p")
        if paragraphs:
            print(f"   ‚û§ Extracted paragraphs: {len(paragraphs)}")
            return "\n\n".join(p.get_text(" ", strip=True) for p in paragraphs), soup

    # Í≥ºÍ±∞ Íµ¨Ï°∞ fallback
    blocks = soup.find_all(["p", "blockquote"])
    if blocks:
        print(f"   ‚û§ Extracted fallback blocks: {len(blocks)}")
        return "\n\n".join(b.get_text(" ", strip=True) for b in blocks), soup

    return soup.get_text(" ", strip=True), soup


# ----------------------------------------------------
# 2) Attendance Ï†úÍ±∞
# ----------------------------------------------------
def remove_attendance_blocks(soup, raw_text):
    cleaned = raw_text
    print("üîµ [CLEAN] Removing attendance blocks...")

    attendance_tag = soup.find("strong", string=lambda x: x and "Attendance" in x)
    if attendance_tag:
        print("   ‚û§ Attendance block found")
        parent = attendance_tag.find_parent("p")

        removed_cnt = 0
        cur = parent
        while cur:
            text = cur.get_text(" ", strip=True)
            if re.search(r"(Chair|President|Economist|Director|Vice|Board)", text) or re.search(r"[A-Z]\.", text):
                cleaned = cleaned.replace(text, "")
                removed_cnt += 1
                cur = cur.find_next("p")
            else:
                break
        print(f"   ‚û§ Removed {removed_cnt} attendance lines")

    # PRESENT block Ï†úÍ±∞
    cleaned = re.sub(r"PRESENT:.*?(?=\n[A-Z])", "", cleaned, flags=re.S)

    # Mr., Ms. Îì± Ìïú Ï§Ñ ÎùºÏù∏ Ï†úÍ±∞
    cleaned = re.sub(r"^(Mr|Ms|Mrs|Messrs)\.[^\n]+$", "", cleaned, flags=re.M)

    cleaned = re.sub(r"\n{2,}", "\n\n", cleaned)
    return cleaned.strip()


# ----------------------------------------------------
# 3) Î¨∏Ïû• Î∂ÑÎ¶¨
# ----------------------------------------------------
def is_garbage_sentence(s):
    if len(s.split()) <= 2:
        return True
    if re.match(r"^[A-Za-z]\.$", s):  # A. / B.
        return True
    if re.match(r"^\d+\.$", s):       # 1. / 2.
        return True
    return False


def smart_sentence_split(text):
    print("üîµ [SPLIT] Splitting sentences...")

    paragraphs = text.split("\n\n")
    result = []

    for para in paragraphs:
        if len(para.split()) < 5:
            continue
        for sent in sent_tokenize(para):
            sent = sent.strip()
            if not is_garbage_sentence(sent):
                result.append(sent)

    print(f"   ‚û§ Total sentences: {len(result)}")
    return result


# ----------------------------------------------------
# 4) ÌÜ†ÌÅ∞Ìôî + POS
# ----------------------------------------------------
def preprocess_sentence(sentence):
    cleaned = re.sub(r"[^a-zA-Z0-9\s]", " ", sentence.lower())
    doc = nlp(cleaned)

    tokens = [t.text for t in doc]
    pos_tags = [t.pos_ for t in doc]

    tokens_no_sw = [t for t in tokens if t not in stop_words]
    cleaned_text = " ".join(tokens_no_sw)

    return tokens, pos_tags, cleaned_text


# ----------------------------------------------------
# 5) ÏµúÏ¢Ö ÏôÑÏÑ±Îêú SECTION LABELS (Î≥ëÌï© + Í∞ïÌôîÎêú Î≤ÑÏ†Ñ)
# ----------------------------------------------------
SECTION_LABELS = {
    "market_conditions": [
        "market",
        "financial markets",
        "open market",
        "open market operations",
        "soma",
        "developments",
        "developments in financial markets",
    ],

    "staff_reviews": [
        "staff review",
        "staff economic",
        "staff financial",
        "staff outlook",
        "review",

        "review of the economic situation",
        "review of the financial situation",
        "review of economic",
        "review of the economic",
        "review of the financial",

        "economic situation",
        "financial situation",
    ],

    "participant_views": [
        "participants' views",
        "participants view",
        "participants",

        "current conditions",
        "economic outlook",
        "current economic conditions and the economic outlook",
        "current conditions and the economic outlook",
        "views on current conditions",
        "views on the economic outlook",
    ],

    "policy": [
        "policy actions",
        "policy action",
        "committee policy",
        "vote",
        "voting",
        "notation vote",
    ]
}


# ----------------------------------------------------
# 6) Ìó§Îçî Ï∂îÏ∂ú
# ----------------------------------------------------
def extract_headers(soup):
    tags = soup.find_all(["h2", "h3", "strong"])
    headers = []

    for t in tags:
        text = t.get_text(" ", strip=True)
        if 1 < len(text.split()) <= 20:
            headers.append(text)

    print(f"üîµ [HEADER] Extracted header count: {len(headers)}")
    return headers


# ----------------------------------------------------
# 7) Ìó§Îçî Í∏∞Î∞ò Section Îß§Ïπ≠ + Î¨∏Ïû• Í∏∞Î∞ò Fallback
# ----------------------------------------------------
def match_section(sentence, headers):

    # Ìó§Îçî Í∏∞Î∞ò Ïö∞ÏÑ† Îß§Ïπ≠
    for hdr in headers:
        h = hdr.lower()
        for label, keywords in SECTION_LABELS.items():
            for kw in keywords:
                if kw in h:
                    return label

    # üî• Fallback: Î¨∏Ïû• Í∏∞Î∞ò Í≤ÄÏÉâ
    s = sentence.lower()

    for label, keywords in SECTION_LABELS.items():
        for kw in keywords:
            if kw in s:
                return label

    return "unknown"


# ----------------------------------------------------
# 8) Ï†ÑÏ≤¥ ÌååÏù¥ÌîÑÎùºÏù∏ Ïã§Ìñâ
# ----------------------------------------------------
def process_urls(urls, output_csv):

    rows = []

    for url in urls:
        print("\n======================================")
        print(f"üöÄ PROCESSING {url}")

        m = re.search(r"(\d{8})", url)
        date = m.group(1)
        doc_id = f"FOMC_{date}"

        raw_text, soup = fetch_minutes_text(url)
        clean_text = remove_attendance_blocks(soup, raw_text)
        sentences = smart_sentence_split(clean_text)
        headers = extract_headers(soup)

        for sent_id, sent in enumerate(sentences):
            tokens, pos_tags, cleaned = preprocess_sentence(sent)
            section = match_section(sent, headers)

            rows.append({
                "doc_id": doc_id,
                "doc_type": "minutes",
                "date": date,
                "title": "FOMC Minutes",
                "source": "Federal Reserve",
                "language": "en",

                "sent_id": sent_id,
                "text": sent,
                "tokens": json.dumps(tokens, ensure_ascii=False),
                "pos_tags": json.dumps(pos_tags, ensure_ascii=False),
                "cleaned_text": cleaned,
                "section": section
            })

            # Ïïû 3Í∞úÏùò Î¨∏Ïû•ÏùÄ ÏÉòÌîå Ï∂úÎ†•
            if sent_id < 3:
                print(f"   ‚û§ Sentence[{sent_id}]: {sent}")
                print(f"      Label = {section}")

    df = pd.DataFrame(rows)
    df.to_csv(output_csv, index=False, encoding="utf-8-sig")

    print("\nüéâ DONE!")
    print("Saved ‚Üí", output_csv)
    print("Total sentences:", len(df))


# ----------------------------------------------------
# Ïã§Ìñâ
# ----------------------------------------------------
# urls = [...]
process_urls(urls, "FOMC_full_sentence_level_with_sections_3.csv")
```
