```python
import pandas as pd
import pandas_datareader.data as web
import datetime

# ==============================================================================
# 1. ì„¤ì •: ê¸°ê°„ ë° ê°€ì ¸ì˜¬ ë°ì´í„° ì½”ë“œ (FRED)
# ==============================================================================
start_date = datetime.datetime(1993, 1, 1)
end_date = datetime.datetime(2025, 12, 31)

# FRED ë°ì´í„° ì½”ë“œ (ë¯¸êµ­ ê¸°ì¤€)
indicators = {
    'FEDFUNDS': 'Base_Rate',  # ê¸°ì¤€ê¸ˆë¦¬ (Federal Funds Effective Rate)
    'CPIAUCSL': 'CPI',        # ì†Œë¹„ìë¬¼ê°€ì§€ìˆ˜ (Consumer Price Index)
    'INDPRO': 'IP'            # ì‚°ì—…ìƒì‚°ì§€ìˆ˜ (Industrial Production)
}

print("ğŸ“¥ FREDì—ì„œ ë¯¸êµ­ ê²½ì œ ë°ì´í„°ë¥¼ ë‹¤ìš´ë¡œë“œí•©ë‹ˆë‹¤...")

# ==============================================================================
# 2. ë°ì´í„° ë‹¤ìš´ë¡œë“œ (ìë™)
# ==============================================================================
try:
    # web.DataReaderë¥¼ ì“°ë©´ FRED ì‚¬ì´íŠ¸ì—ì„œ ì•Œì•„ì„œ ê¸ì–´ì˜µë‹ˆë‹¤.
    df_macro = web.DataReader(list(indicators.keys()), 'fred', start_date, end_date)

    # ì»¬ëŸ¼ ì´ë¦„ ë³´ê¸° ì¢‹ê²Œ ë³€ê²½ (FEDFUNDS -> Base_Rate ë“±)
    df_macro = df_macro.rename(columns=indicators)

    # ë‚ ì§œ ì¸ë±ìŠ¤ë¥¼ ì»¬ëŸ¼ìœ¼ë¡œ êº¼ë‚´ê¸°
    df_macro = df_macro.reset_index()

    # ë‚ ì§œ í˜•ì‹ í†µì¼ (YYYY-MM-01) - ì›”ì´ˆ ê¸°ì¤€
    df_macro['DATE'] = df_macro['DATE'].dt.to_period('M').dt.to_timestamp()

    # ==============================================================================
    # 3. íŒŒì¼ ì €ì¥ (macro_data.csv)
    # ==============================================================================
    output_file = "macro_data.csv"
    df_macro.to_csv(output_file, index=False)

    print(f"\nâœ… ì„±ê³µ! ë°ì´í„°ê°€ '{output_file}'ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
    print(f"   - ê¸°ê°„: {start_date.year} ~ {end_date.year}")
    print(f"   - ë°ì´í„° ê°œìˆ˜: {len(df_macro)}ê°œì›”")
    print("\n=== ë°ì´í„° ë¯¸ë¦¬ë³´ê¸° ===")
    print(df_macro.head())
    print("...")
    print(df_macro.tail())

except Exception as e:
    print(f"\nâŒ ë‹¤ìš´ë¡œë“œ ì‹¤íŒ¨: {e}")
    print("íŒ: 'pip install pandas_datareader'ë¥¼ ì„¤ì¹˜í–ˆëŠ”ì§€ í™•ì¸í•´ë³´ì„¸ìš”.")
```
